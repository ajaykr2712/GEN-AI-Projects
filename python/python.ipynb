```python
import matplotlib.pyplot as plt


x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]


fig, ax = plt.subplots()


ax.plot(x, y, label='Linear growth')
ax.scatter(x, y, label='Linear growth', color='pink')

ax.set_title('Simple Line Plot')


ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')


ax.legend()

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data for plotting
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Create a figure and an axes.
fig, ax = plt.subplots()

# Plotting the bar chart
ax.bar(x, y, color='green', label='Bar heights')

# Adding a title
ax.set_title('Simple Bar Chart')

# Adding x and y labels
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')

# Adding a legend
ax.legend()

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data for plotting
labels = ['One', 'Two', 'Three', 'Four', 'Five']
sizes = [2, 3, 5, 7, 11]

# Create a pie chart
fig, ax = plt.subplots()
ax.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)

# Equal aspect ratio ensures that pie is drawn as a circle.
ax.axis('equal')

# Adding a title
ax.set_title('Simple Pie Chart')

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for plotting
data = np.random.normal(5, 2, 1000)

# Create a figure and an axes.
fig, ax = plt.subplots()

# Plotting the histogram
ax.hist(data, bins=30, color='purple', label='Data distribution')

# Adding a title
ax.set_title('Simple Histogram')

# Adding x and y labels
ax.set_xlabel('Value')
ax.set_ylabel('Frequency')

# Adding a legend
ax.legend()

# Display the plot
plt.show()
```
